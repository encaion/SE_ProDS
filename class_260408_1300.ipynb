* .groupby(): https://datadoctorblog.com/2023/04/28/Py-Basic-Pandas-Groupby/

```python
df = pd.read_csv("diamonds.csv")
df.head(2)
```

```python
df_g1 = df.groupby("cut")["depth"].max().reset_index()
df_g1
```

```python
df_g2 = df.groupby(["cut", "color"])["depth"].max().reset_index()
df_g2 = df_g2.sort_values(["color", "cut"], ascending = [False, True])
df_g2.head()
```

```python
df_g3 = df.groupby("color")[["price", "carat"]].min().reset_index()
df_g3
```

```python
df_g4 = df.groupby("color")["price"].agg(["min", "max"]).reset_index()
df_g4
```

```python
df_g5 = df.groupby("color")[["price", "carat"]].agg({"price": "min",
                                                     "carat": "max"}).reset_index()
df_g5 # 🎯🎯시험 기출 전처리 유형🎯🎯
```

```python
df_bike = pd.read_csv("bike.csv")
df_dt2 = df_bike[["datetime"]].copy()
```

```python
df_dt2["datetime"] = pd.to_datetime(df_dt2["datetime"])
df_dt2["date"  ] = df_dt2["datetime"].dt.date
df_dt2["yyyymm"] = df_dt2["datetime"].dt.strftime("%Y-%m") # 기출 🎯
df_dt2["year"  ] = df_dt2["datetime"].dt.year
df_dt2["month" ] = df_dt2["datetime"].dt.month
df_dt2["wday"  ] = df_dt2["datetime"].dt.weekday
df_dt2["wend"  ] = np.where(df_dt2["wday"] >= 5, 1, 0)
df_dt2["hour"  ] = df_dt2["datetime"].dt.hour
df_dt2["dt_s9h"] = df_dt2["datetime"] + pd.to_timedelta(9, unit = "h")
```

```python
df_t = pd.DataFrame(dict(date = pd.date_range("2026-04-08", "2026-04-14"),
                         v = [100, 200, 300, np.nan, 500, 600, 700]))
```

```python
df_t["v_s1"] = df_t["v"].shift(1)
df_t["v_sn1"] = df_t["v"].shift(-1)
df_t["v_fill_1"] = np.where(df_t["v"].isna(), 
                            (df_t["v_s1"] + df_t["v_sn1"]) / 2, 
                            df_t["v"])
df_t["v_fill_2"] = df_t["v"].interpolate()
```
