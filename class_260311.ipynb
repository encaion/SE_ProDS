```python
df_dt2["datetime"] = pd.to_datetime(df_dt2["datetime"])
df_dt2["date"  ] = df_dt2["datetime"].dt.date
df_dt2["yyyymm"] = df_dt2["datetime"].dt.strftime("%Y-%m") # 기출 🎯
df_dt2["year"  ] = df_dt2["datetime"].dt.year
df_dt2["month" ] = df_dt2["datetime"].dt.month
df_dt2["wday"  ] = df_dt2["datetime"].dt.weekday
df_dt2["wend"  ] = np.where(df_dt2["wday"] >= 5, 1, 0) # 주말 1, 주중 0
df_dt2["hour"  ] = df_dt2["datetime"].dt.hour
df_dt2["dt_s9h"] = df_dt2["datetime"] + pd.to_timedelta(9, unit = "h")
```

```python
df_e = pd.read_csv("elec_load_2017_7d.csv")
df_e.head(2)
```

```python
df_e_m1 = df_e.melt(id_vars = ["YEAR", "MONTH", "DAY"],
                    var_name = "HOUR", value_name = "LOAD") 
df_e_m1.head()
```

```python
df_e_m1["h1"] = df_e_m1["HOUR"].str.replace("X", "").str.replace("HR", "")
df_e_m1["h2"] = df_e_m1["HOUR"].str.replace("X|HR", "", regex = True) # X 또는 HR
df_e_m1["h3"] = df_e_m1["HOUR"].str.replace("[A-Z]", "", regex = True) # 영문 대문자
df_e_m1["h4"] = df_e_m1["HOUR"].str.replace("[^0-9]", "", regex = True) # 숫자를 제외한 나머지
# regex: regular expression(정규표현식)
df_e_m1.tail()
```

```python
df_ck1 = pd.crosstab(index = df["cut"], columns = df["color"], values = df["price"],
                     aggfunc = lambda x: round(x.mean()))
df_ck1
```

```python
def udf_mean_r(x):
    return round(x.mean())
    
df_ck2 = df.groupby(["cut", "color"])["price"].agg(udf_mean_r).reset_index()
df_ck2.head()
```
